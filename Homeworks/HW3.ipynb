##Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

##Preparing Data

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100, n_features=8, random_state=42, n_informative = 5, class_sep = 2)
```

```python
data = pd.DataFrame(X,y)
```

```python
data.columns =['Col_0','Col_1', 'Col_2','Col_3','Col_4','Col_5','Col_6','Col_7']
```

```python
data.reset_index(drop = False)
```

##Outlier Detection

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(data))
z
```

```python
outliers = list(set(np.where(z > 3)[0]))

len(outliers)
```

##Spliting dataset into train and test set

```python
data.head()
```

```python
new_data=data.reset_index(drop = False)
```

```python
X_new = new_data.drop('index', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
X_scaled = MinMaxScaler().fit_transform(X_new)
X_scaled
```

```python
from sklearn.model_selection import train_test_split, cross_validate
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.3, random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
models = LogisticRegression(random_state=42)
cv = cross_validate(models,X_train,y_train, cv = 3, return_estimator=True)
```

```python
print("Mean training accuracy: {}".format(np.mean(cv['test_score'])))
print("Test accuracy: {}".format(cv["estimator"][0].score(X_test,y_test)))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pred = cv["estimator"][0].predict(X_test)

cm = confusion_matrix(y_test, pred)
```

```python
plt.figure(figsize=(12, 8))
ax =sns.heatmap(cm, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(["False","True"], fontsize = 12)
ax.yaxis.set_ticklabels(["False","True"], fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, cv["estimator"][0].predict(X_test)))
```

```python
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score

final_model = cv["estimator"][0]

y_pred = final_model.predict(X_test)

print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

##Desicion Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train,y_train)
print("Accuracy of train:",clf.score(X_train,y_train))
print("Accuracy of test:",clf.score(X_test,y_test))
```

```python
cv = cross_validate(clf,X_train,y_train, cv = 3, return_estimator=True)
```

```python
pred1 = cv["estimator"][0].predict(X_test)
cm1 = confusion_matrix(y_test, pred1)
```

```python
plt.figure(figsize=(12, 8))
ax =sns.heatmap(cm1, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(["False","True"], fontsize = 12)
ax.yaxis.set_ticklabels(["False","True"], fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score
```

```python
print("Precision = {}".format(precision_score(y_test, pred1, average='macro')))
print("Recall = {}".format(recall_score(y_test, pred1, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, pred1)))
print("F1 Score = {}".format(f1_score(y_test, pred1,average='macro')))
```

##XGBoost

```python
import xgboost as xgb
```

```python
dmatrix_train = xgb.DMatrix(data=X_train, label=y_train)
dmatrix_test = xgb.DMatrix(data=X_test, label=y_test)
```

```python
param = {'max_depth':3, 
         'eta':1, 
         'objective':'multi:softprob', 
         'num_class':3}

num_round = 5
model = xgb.train(param, dmatrix_train, num_round)
```

```python
preds = model.predict(dmatrix_test)
preds[:10]
```

```python
best_preds = np.asarray([np.argmax(line) for line in preds])
best_preds
```

```python
print("Precision = {}".format(precision_score(y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(y_test, best_preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, best_preds)))
print("F1 Score = {}".format(f1_score(y_test, best_preds,average='macro')))
```

```python
plt.figure(figsize=(12, 8))
confm = confusion_matrix(y_test, best_preds)
ax = sns.heatmap(confm, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(["False","True"], fontsize = 12)
ax.yaxis.set_ticklabels(["False","True"], fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```
